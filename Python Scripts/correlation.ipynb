```python
def calculate(i,col):
  # DO (mg/L)
  if(col== 1):
    return ((i-14)*100/9)*0.3723
  #ph
  if(col==2):
    return ((i-7)*100/1.5)*0.1176
  #Electrical conductivity
  if(col==3):
    return ((i)/300)*0.0033
  #BOD (mg/L)
  if(col==4):
    return (i*100/5)*0.3723
  #Turbidity (NTU)
  if(col==5):
    return (i*100/5)*0.1
  #COD (mg/L)
  if(col==6):
    return (i*100/10)*0.12
  #Hardness (mg/L)
  if(col==7):
    return (i*100/300)*0.0033
  #TDS (mg/L)
  if(col==8):
    return (i*100/1000)*0.001
  #TSS (mg/L)
  else:
    return (i*100/500)*0.0037



```

```python
import csv
import pandas as pd

def read_csv_rows(csv_file_path):
    rows_data = []
    with open(csv_file_path, mode='r') as file:
        reader = csv.reader(file)
        for row in reader:
            rows_data.append(row)
    return rows_data

# Example usage:
csv_file_path = '/content/drive/MyDrive/Basara.csv'  # Update with your CSV file path
rows_data = read_csv_rows(csv_file_path)

# Define your calculate function here (assuming you have defined it elsewhere in your code)

# Create a new column for WQI
rows_data[0].append('WQI')  # Assuming the first row contains column headers

# Calculate and assign WQI for each row
num_columns = 10
for row in range(1, len(rows_data)):
    wqi = 0
    for col in range(1, num_columns - 1):
        i = float(rows_data[row][col])  # Convert string to float
        wn = calculate(i, col)  # Assuming you have defined the calculate function elsewhere
        wqi += wn
    rows_data[row].append(wqi)

# Convert the list of rows to a DataFrame
df = pd.DataFrame(rows_data[1:], columns=rows_data[0])

# Print the DataFrame with the calculated WQI column
print(df)
```

```python

from sklearn.preprocessing import MinMaxScaler
columns_to_keep = df.columns[1:-2]
#using MinMax Scalar
scaler = MinMaxScaler()

df[columns_to_keep] = scaler.fit_transform(df[columns_to_keep])

# values between 0 and 1
print(df.head())
```

```python

import pandas as pd


# Compute the correlation matrix
correlation_matrix = df.corr()

# Get correlations with the target variable (assuming 'WQI' is your target variable)
target_correlations = correlation_matrix['WQI'].abs().sort_values(ascending=False)

# Print the most correlated features
print(target_correlations)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Separate features (X) and target variable (y)
X = df.drop(['WQI'], axis=1)  # Assuming 'WQI' is your target variable
y = df['WQI']

# Train a Random Forest model
rf_model = RandomForestRegressor()
rf_model.fit(X, y)

# Get feature importances
feature_importances = rf_model.feature_importances_

# Create a DataFrame to store feature importances
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
})

# Sort features by importance
sorted_features = feature_importance_df.sort_values(by='Importance', ascending=False)

# Print the most important features
print(sorted_features)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression  # Example model (use appropriate model for your data)

# Initialize the model (you can use any suitable model)
model = LinearRegression()

# Initialize RFE and specify the number of features to select
rfe = RFE(model, n_features_to_select=5)  # Select top 5 features (adjust as needed)

# Fit RFE
rfe.fit(X, y)

# Get selected features
selected_features = X.columns[rfe.support_]

# Print the selected features
print(selected_features)
```
